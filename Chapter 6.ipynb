```python
tips = sns.load_dataset("tips")
tips
```

```python
group_tips = tips.groupby("sex", observed=True).agg(
    avg_tips = ("tip", "mean"),
    sum_bill = ("total_bill", "sum")
).reset_index().rename(columns={"sex": "gender"})

group_tips
```

```python
group_tips.plot.pie(x="gender", y="sum_bill", labels=group_tips["gender"], title="Total Bill Distribution between Gender")
```

```python
import matplotlib.pyplot as plt
plt.figure()

# Plot with labels
plt.plot(iris.sepal_length, label='Sepal Length')
plt.plot(iris.petal_length, label='Petal Length')

# Add title and labels
plt.title('Comparison of Sepal and Petal Lengths in Iris Dataset', pad=20)
plt.xlabel('Observation Index')
plt.ylabel('Length (cm)')

# Add legend and grid
plt.legend()

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
